```python
import numpy as np
import statistics as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age = [10,15,20,25,30,35,40,45,50, 170, 190]
```

```python
len(age)
```

```python
q1=np.quantile(age,0.25)
q1
```

```python
q3=np.quantile(age,0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
ub=q3+1.5*IQR
ub
```

```python
lb=q1-1.5*IQR
lb
```

```python
out=[]
for i in age:
  if (i>=lb or i<=ub):
    out.append(i)
  else:
    print(i)
```
