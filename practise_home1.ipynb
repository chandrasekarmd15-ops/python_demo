```python
class vehicle:
  def __init__(self,color,brand,seaters):
    self.color=color
    self.brand=brand
    self.seaters=seaters
  def displayinfo(self):
    print("color :",self.color)
    print("brand :",self.brand)
    print("seaters :",self.seaters)
class car(vehicle):
  def __init__(self,color,brand,seaters,mileage,fuel):
    super().__init__(color,brand,seaters)
    self.mileage=mileage
    self.fuel=fuel
  def display(self):
    super().displayinfo()
    print("mileage  ::  ",self.mileage)
    print("fuel used  ::  ",self.fuel)
class electric(car):
  def __init__(self,color,brand,seaters,mileage,fuel,drive_mode):
    super().__init__(color,brand,seaters,mileage,fuel)
    self.drive_mode=drive_mode
  def show(self):
    super().display()
    print("drive mode in electric car is  ::  ",self.drive_mode)
```

```python
obj1=car("BLACK","AUDI",2,10,"PETROL")
```

```python
obj1.display()
```

```python
obj2=electric("WHITE","BMW",5,300,"ELECTRIC","AUTOMATIC")
```

```python
obj2.show()
```

```python
##  Build a simple calculator class.
##  Include methods for basic arithmetic operations like add, subtract, multiply, and divide.
## for 2  different numbers for each method  ::

class calculater():
  def add(self,no1,no2):
    print("ADD    ::    ",no1+no2)
  def sub(self,no1,no2):
    print("SUB    ::    ",no1-no2)
  def mult(self,no1,no2):
    print("MULT    ::    ",no1*no2)
  def div(self,no1,no2):
    print("DIV    ::    ",no1/no2)
```
